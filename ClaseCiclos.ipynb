# Ciclo While
    Mientras que la expresión lógica sea igual a True, entonces, el ciclo se desarrolla
    indefinidamente hasta que la expresión lógica sea igual a False

    Sintaxis:

    while -condición-:
        bloque de codigo
        si condición es
        True

    Bloque de código fuera del ciclo
 

```python
Contar = 1

while Contar <= 2000:
    print("Sebastian",Contar)
    Contar+=1
```

# Ejercicio 1
    Imprima su nombre del 1 al 30 y su apellido del 31 al 60
    y su nombre nuevamente del 61 al 100

```python
Contador = 1
Nombre = "Juan Sebastian"
Apellido = "Pastrana Ardila"
while Contador <= 100:
    if Contador>=1 and Contador <=30:
        print(Nombre,Contador)
    elif Contador>30 and Contador <=61:
        print(Apellido,Contador)
    else:
        print(Nombre,Contador)
    Contador+=1
```

# Ejercicio 2

    Hacer un programa que captura la contraseña y en caso de no ser válida, debe mantener preguntando
    indefinidamente

```python
Ciclo = True
while Ciclo == True:
    ContrasenaSecreta = "ContraseñaSecreta"
    Contrasena = input("Ingrese la contraseña: ")
    if Contrasena == ContrasenaSecreta:
        print("\nContraseña correcta!")
        Ciclo = False
    else:
        print("\nContraseña incorrecta, intente nuevamente")
```

# Calcular el factoria de cualquier número

```python
Numero = input("Ingrese número a calcular factorial: ")
Contador = 1
Respuesta = 1
try:
    Numero = int(Numero)
    while Contador<=Numero:
        Respuesta = Respuesta*Contador
        Contador+=1
    print("Factorial:",Respuesta)
    
except ValueError:
    print("Error, no ingresó un numero entero")
```
